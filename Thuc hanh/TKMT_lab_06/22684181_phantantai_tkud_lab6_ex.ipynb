### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:

a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
POP = np.random.uniform(120, 200, 10000)

POP_mean = np.mean(POP)
POP_std = np.std(POP)
print("Trung bình của quần thể", POP_mean)
print("Độ lệch chuẩn về chiều cao của quần thể", POP_std)
```

```python
high_person  = []
for i in POP:
    if i >= 180:
        high_person.append(i)
        
high_person = np.array(high_person)
percent_high_person = len(high_person) / 1000

print("tỷ lệ người cao trong quần thể", percent_high_person)
```

```python
plt.hist(POP, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram về chiều cao của quần thể")
plt.xlabel("Chiều cao")
plt.ylabel("Số lượng")
plt.grid(True)
plt.show()
```

Quần thể không phân phối chuẩn !

```python
plt.hist(POP, bins=30, edgecolor='black', alpha=0.6, density=True, label='Quần thể')

x = np.linspace(min(POP), max(POP), 1000)
plt.plot(x, norm.pdf(x, loc=POP_mean, scale=POP_std), 'r', label='Phân phối chuẩn')
plt.title('Histogram và hàm mật độ của phân phối chuẩn')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Mật độ xác suất')
plt.legend()
plt.grid(True)
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


Công thức tính khoảng tin cậy với trường hợp biết $\sigma$

$\text{CI} = \left[ \bar{x} - z_{\frac{\alpha}{2}} \cdot \frac{\sigma}{\sqrt{n}}, \bar{x} + z_{\frac{\alpha}{2}} \cdot \frac{\sigma}{\sqrt{n}} \right]$

```python
can_nang = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])
mean_can_nang = np.mean(can_nang)
sigma = 0.3

# Giá trị tới hạn z cho độ tin cậy 95% (z = 1.96)
z_value = norm.ppf(0.975)

margin_of_error = z_value * (sigma / np.sqrt(len(can_nang)))
lower_bound = mean_can_nang - margin_of_error
upper_bound = mean_can_nang + margin_of_error

print(f"Khoảng tin cậy 95% cho cân nặng trung bình: [{lower_bound:.4f}, {upper_bound:.4f}] kg")
```

Công thức tính khoảng tin cậy với trường hợp không biết $\sigma$

trong trường hợp không biết độ lệch chuẩn của tổng thể 
𝜎, chúng ta sẽ sử dụng phân phối t-Student (hay t-distribution) để lập khoảng tin cậy cho trung bình, thay vì sử dụng phân phối chuẩn.

$\text{CI} = \left[ \bar{x} - t_{\frac{\alpha}{2}, n-1} \cdot \frac{s}{\sqrt{n}}, \bar{x} + t_{\frac{\alpha}{2}, n-1} \cdot \frac{s}{\sqrt{n}} \right]$

với $t_{\frac{\alpha}{2}, n-1}$ là giá trị tới hạn từ phân phối t-Student với mức độ tin cậy và 
𝑛
−
1
n−1 bậc tự do (với độ tin cậy 95%).


```python
from scipy.stats import t

std_weight = np.std(can_nang, ddof=1)  # Độ lệch chuẩn mẫu (ddof=1 để tính s)
# Giá trị tới hạn t cho độ tin cậy 95% với n-1 bậc tự do
t_value = t.ppf(0.975, df=len(can_nang)-1)

margin_of_error = t_value * (std_weight / np.sqrt(len(can_nang)))
lower_bound = mean_can_nang - margin_of_error
upper_bound = mean_can_nang + margin_of_error

print(f"Trung bình mẫu: {mean_can_nang:.4f} kg")
print(f"Độ lệch chuẩn mẫu: {std_weight:.4f} kg")
print(f"Khoảng tin cậy 95% cho cân nặng trung bình: [{lower_bound:.4f}, {upper_bound:.4f}] kg")
```

Bộ dữ liệu `SAT-lard.csv` chứa điểm thi SAT của 1000 học sinh. 

a. Giả sử bộ dữ liệu là kết quả của quá trình lấy mẫu ngẫu nhiên từ các học sinh. Bạn hãy xây dựng khoảng tin cậy cho trung bình điểm thi của tất cả các sinh viên với độ tin cậy 99% (`Đáp án: 105.2 ± 3.10`)

c. Giả sử bộ dữ liệu là toàn bộ kết quả điểm thi SAT của học sinh một trường phố thông. Tính điểm SAT trung bình của trường phổ thông đấy. (`Đáp án:  μ = 1528.74`)

d. Giả sử bộ dữ liệu là toàn bộ kết quả điểm thi SAT của học sinh một trường phố thông. Sử dụng mẫu là điểm thi của 36 học sinh đầu tiên trong bộ dữ liệu, bạn hãy xây dựng khoảng tin cậy **99%** cho điểm SAT trung bình của trường phổ thông đấy. (`Đáp án: (1428.22, 1602.89)`)

e. Giả sử bộ dữ liệu là toàn bộ kết quả điểm thi SAT của học sinh một trường phố thông. Sử dụng mẫu là điểm thi của 36 học sinh đầu tiên trong bộ dữ liệu, bạn hãy xây dựng khoảng tin cậy **95%** cho điểm SAT trung bình của trường phổ thông đấy. Cho biết khoảng ước lượng bạn xây dựng có đúng không, dựa trên giá trị đã tính ở câu trên.

```python
import pandas as pd
import numpy as np
from scipy import stats

data = pd.read_csv('SAT-lard.csv')

sat_scores = data['SAT Score']

mean_score = np.mean(sat_scores)
std_dev = np.std(sat_scores, ddof=1)  # ddof=1 để tính độ lệch chuẩn mẫu

n = len(sat_scores)

# Tính khoảng tin cậy 99%
confidence_level = 0.99
z_score = stats.norm.ppf((1 + confidence_level) / 2)

margin_of_error = z_score * (std_dev / np.sqrt(n))
confidence_interval = (mean_score - margin_of_error, mean_score + margin_of_error)

print(f"Trung bình điểm thi: {mean_score:.2f}")
print(f"Khoảng tin cậy 99%: {confidence_interval[0]:.2f} đến {confidence_interval[1]:.2f}")
```

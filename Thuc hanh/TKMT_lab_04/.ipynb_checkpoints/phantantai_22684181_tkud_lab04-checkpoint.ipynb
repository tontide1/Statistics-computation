# KHÁM PHÁ DỮ LIỆU

Dữ liệu chứa trong file `tips.csv` chứa thông tin tiền tip của khách hàng tại một nhà hàng. Dữ liệu có các thông tin sau:

- `total_bill`: tổng hoá đơn (USD)

```python
#HIỂU DỮ LIỆU
df = pd.read_csv('tips.csv')

# 2 Xoá cột unnamed:0
df = df.drop("Unnamed: 0", axis='columns')
df.head()
```

```python
#3
print('Số dòng của tệp dữ liệu: ', df.shape[0])
print('Số thuộc tính của tệp dữ liệu: ', df.shape[1])
```

```python
#4
df.info()
```

Từ ouput của `df.info()` ta dễ dàng thấy:

- Thuộc tính định lượng : `total_bill` `tip` `size`

- Thuộc tính phân loại: `sex` `smoker` `day` `time`

```python
#5
count_sex = df['sex'].value_counts()
total_sex = len(df['sex'])
print(count_sex)

print("Tỷ lệ giữa nam và nữ: \n", count_sex/total_sex * 100)
```

```python
#6
count_smoker = df['smoker'].value_counts()
total_smoker = len(df['smoker'])
per_smoker = count_smoker/total_smoker * 100
print("Số người hút thuốc: ", count_smoker[1])
print("Tỷ lệ người hút thuốc: ", per_smoker[1])
```

```python
#7
df['day'].describe()
```

```python
per_day = 87 / 244 * 100
print("Tỷ lệ ngày trong tuần khách hàng đến ăn:", per_day)
```

Từ output của `df['day'].describe()` ta thấy:

- Sat (thứ 7) là ngày trong tuần mà khách ghé ăn nhiều nhất

```python
#8
df['time'].describe()
```

Từ output của `df['time'].describe()` ta thấy:

- Buổi tối (dinner) là buổi mà khách ghé ăn nhiều nhất

```python
#9 và 10
df['size'].describe()
```

Từ output của `df['size'].describe()` ta thấy:

- Khách dùng bữa nhiều nhất là 6 người

- Trung bình 1 bữa ăn phục vụ 2.57 khách

```python
#YẾU TỐ ẢNH HƯỞN ĐẾN TIỀN TIP
so_luong_nhan_duoc_tip=len(df[df['tip'] > 0])
so_luong_tip = len(df['tip'])
print("Tỷ lệ nhận được tiền tip của nhà hàng", so_luong_nhan_duoc_tip/ so_luong_tip * 100, '%')
```

```python
#2
total_tip = df['tip'].sum()
total_bill = df['total_bill'].sum()

tip_percentage = (total_tip / total_bill) * 100

print("Tỷ lệ tiền tip trung bình:", tip_percentage, "%")
```

```python
#3
sum_tip = df.groupby('sex')['tip'].sum()
sum_tip
```

Từ output trên ta thấy `Male` có tổng số tiền tip nhiều hơn

```python
count_tip_sex = df.groupby('sex')['tip'].count()
count_tip_sex
```

Từ output trên ta thấy `Male` có số lần được tip nhiều hơn

```python
#4
sum_smoker_tip = df.groupby('smoker')['tip'].sum()
sum_smoker_tip
```

Từ ouput trên ta thấy `Không hút thuốc` được tip nhiều hơn

```python
count_smoker_tip = df.groupby('smoker')['tip'].count()
count_smoker_tip
```

Từ ouput trên ta thấy `Không hút thuốc` được tần suất tip nhiều hơn

```python
#5
sum_tip_per_day = df.groupby('day')['tip'].sum()
sum_tip_per_day
```

```python
#TRỰC QUAN DỮ LIỆU
# Matplotlib
import matplotlib.pyplot as plt
#1
plt.bar(sum_tip_per_day.index, sum_tip_per_day.values)

plt.title('Tổng tiền tip theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Tổng tiền tip')

plt.show()
```

```python
#2
sum_tip_per_time = df.groupby('time')['tip'].sum()
sum_tip_per_time

plt.bar(sum_tip_per_time.index, sum_tip_per_time.values)

plt.title('Tổng tiền tip theo bữa ăn')
plt.xlabel('Bữa ăn')
plt.ylabel('Tổng tiền tip')

plt.show()
```

```python
#3
plt.hist(df['total_bill'])
plt.title('Tổng hoá đơn')
plt.show()
```

```python
#4
plt.boxplot(df['total_bill'])
plt.title('Tổng hoá đơn')
plt.show()
```

```python
#5
plt.scatter(df['total_bill'], df['tip'])
plt.xlabel('Tổng hóa đơn')
plt.ylabel('Tiền tip')
plt.title('Mối quan hệ giữa tổng hóa đơn và tiền tip')
plt.show()
```

```python
#6
plt.bar(df['day'], df['total_bill'])
plt.ylabel('Tổng hóa đơn')
plt.xlabel('Ngay')
plt.title('Mối quan hệ giữa tổng hóa đơn và ngay')
plt.show()
```

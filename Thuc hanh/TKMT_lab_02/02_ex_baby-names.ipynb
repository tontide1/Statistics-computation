### 09. Vẽ đồ thị top 10 tên xuất hiện nhiều nhất năm 2014

```python
import matplotlib.pyplot as plt
```

```python
df.head(10)
```

```python
df_2014 = df[df['Year'] == 2014]
name_counts = df_2014['Name'].value_counts().head(10)

plt.figure(figsize = (10,6))
plt.bar(name_counts.index, name_counts.values)
plt.title('10 tên phổ biến nhất năm 2014')
plt.xlabel('Tên')
plt.ylabel('Số lần xuất hiện')
plt.xticks(rotation = 45)

plt.show()
```

### 10. Vẽ đồ thị thể hiện tỷ lệ nam nữ qua các năm 2004-2014

```python
def cal_gender(year):
    df_current = df[df['Year'] == year]
    m_current = df_current[df_current['Gender'] == 'M']['Gender'].count()
    f_current = df_current[df_current['Gender'] == 'F']['Gender'].count()
    res = round((m_current / f_current) * 100, 2)
    return res
```

```python
gender_2004 = cal_gender(2004)
gender_2005 = cal_gender(2005)
gender_2006 = cal_gender(2006)
gender_2007 = cal_gender(2007)
gender_2008 = cal_gender(2008)
gender_2009 = cal_gender(2009)
gender_2010 = cal_gender(2010)
gender_2011 = cal_gender(2011)
gender_2012 = cal_gender(2012)
gender_2013 = cal_gender(2013)
gender_2014 = cal_gender(2014)


gender_data = [gender_2004, gender_2005, gender_2006, gender_2007,
              gender_2008, gender_2009, gender_2010, gender_2011,
              gender_2012, gender_2013, gender_2014]
year = [2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]
plt.plot(gender_data, year)
plt.title('Tỉ lệ giới tính qua các năm 2004 - 2014')
plt.xlabel('Tỉ lệ giới tính')
plt.ylabel('Năm')
plt.show()
```
